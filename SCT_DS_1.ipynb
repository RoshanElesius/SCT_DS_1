```python
# Load data from CSV and skip metadata rows if necessary
df = pd.read_csv("population.csv", skiprows=3)  # Adjust skiprows as needed
print(df.columns)  # Print columns to verify the correct names

# Use the population data for a specific year, e.g., '2023'
df['2023'] = pd.to_numeric(df['2023'], errors='coerce')

# Exploratory Analysis: Basic stats
population_summary = df['2023'].describe()
print("Population Summary Statistics for 2023:\n", population_summary)

# Visualize: Histogram of Population, Total for 2023
plt.figure(figsize=(10, 6))
plt.hist(df['2023'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Population Distribution by Country in 2023')
plt.xlabel('Population (Total)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Load data from CSV and skip metadata rows if necessary
df = pd.read_csv("population.csv", skiprows=3)  # Adjust skiprows if needed
print(df.columns)  # Check columns to verify structure

# Extract only the population data (columns with years) and country names
# Filter to only include numeric columns (years) and 'Country Name'
years = df.columns[4:-1]  # Assuming years start from 1960 and go up to 2023, adjust as necessary
population_data = df[['Country Name'] + list(years)]

# Convert population data columns to numeric, handling any non-numeric entries
population_data[years] = population_data[years].apply(pd.to_numeric, errors='coerce')

# Set 'Country Name' as the index for easy plotting
population_data.set_index('Country Name', inplace=True)

# Plot: Population trend over the years for each country
plt.figure(figsize=(14, 8))

# Plot each country's population over time
for country in population_data.index:
    plt.plot(years, population_data.loc[country], label=country)

plt.title('Population Trends Over Time by Country')
plt.xlabel('Year')
plt.ylabel('Population (Total)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Position legend outside the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to make room for legend

plt.show()
```
